```python
import csv
import json
import random
from datasets import load_dataset


dataset = load_dataset("uiyunkim-hub/pubmed-abstract", split="train", streaming=True)


dataset_head = dataset.take(5000)

raw_texts = [x['abstract'].replace('\n', ' ').strip() for x in dataset_head]
clean_texts = [t for t in raw_texts if len(t) > 20]


with open("pretrain.txt", "w", encoding="utf-8") as f:
    f.write("\n\n".join(clean_texts))

pos_samples = clean_texts[:250]
random.seed(42)

subjects = ["I", "We", "My friend", "My brother", "My sister", "They", "Someone"]
verbs = ["went", "walked", "called", "texted", "cooked", "watched", "played", "visited", "cleaned", "bought"]
places = ["the park", "the supermarket", "the library", "a cafe", "home", "the gym", "downtown", "the office", "the mall"]
times = ["today", "yesterday", "this morning", "last night", "after class", "on the weekend"]
extras = [
    "and it was pretty chill.",
    "because I was bored.",
    "and then I took a nap.",
    "and it took forever.",
    "and I forgot my keys again.",
    "and it was actually fun."
]

neg_set = set()
while len(neg_set) < 250:
    s = f"{random.choice(subjects)} {random.choice(verbs)} to {random.choice(places)} {random.choice(times)} {random.choice(extras)}"
    neg_set.add(s)

neg_samples = list(neg_set) #First I was generating negative by multiplying the same 10 sentences until there was 250 of them. AI suggested this change to increase variety.


csv_data = [[t, 1] for t in pos_samples] + [[t, 0] for t in neg_samples]
random.shuffle(csv_data)

with open("classification.csv", "w", newline='', encoding="utf-8") as f:
    writer = csv.writer(f, quoting=csv.QUOTE_ALL) #AI suggested adding quoting for better CSV formatting
    writer.writerow(["text", "label"])
    writer.writerows(csv_data)

json_data = [{"instruction": "Medical Question Placeholder", "response": "Medical Answer Placeholder"} for _ in range(50)]

with open("instruct.json", "w", encoding="utf-8") as f:
    json.dump(json_data, f, indent=4)

```
